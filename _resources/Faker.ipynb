```python
def generate_user():
    prn = "103222" + str(fake.unique.random_number(digits=4))  # 10-digit PRN
    username = fake.user_name()
    password = fake.password()  # Generate a random password
    hashed_password = bcrypt.hashpw(password.encode('utf-8'), bcrypt.gensalt())  # Hash the password

    # Generate last 5 tags of a subject (assuming you have a predefined list of tags)
    last_5_tags = sample(tags, 5)  # Get last 5 tags

    user_data = {
        "prn": prn,
        "username": username,
        "password": hashed_password.decode('utf-8'),  # Store as string
        "last_5_tags": last_5_tags
    }

    return user_data
```

```python
for _ in range(5):
    user = generate_user()
    collection.insert_one(user)
```
